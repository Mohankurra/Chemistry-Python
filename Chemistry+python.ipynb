```python
#Number of elements, what elements and the percentage of elements

elements = []
percentages = []

atomic_masses = {'H':1, 'He':4, 'Li':7, 'Be':9, 'B':11, 'C':12, 'N':14, 'O':16, 'F':19, 'Cl':35.5}

n = int(input('Number of different elements: '))

order = 1
for i in range(n):
  ele = str(input(f'the element number {order}: '))
  elements.append(ele)
  proportion = float(input(f'the percentage of element {ele} in the molecule: '))
  percentages.append(proportion)
  order +=1

```

```python
print(elements)
```

```python
percentages
```

```python
ratios = []

count = 0

while count < len(elements):
  ratio = percentages[count]/atomic_masses[elements[count]]
  ratios.append(ratio)
  count +=1
```

```python
ratios
```

```python
least = min(ratios)

least_integer_coefs = []

for i in ratios:
  least_int = i/least
  least_integer_coefs.append(least_int)

```

```python
print(least_integer_coefs)
```

```python
#Determining the least integer ratios

integer = 1
while True:
  score = 0
  for i in least_integer_coefs:
      score +=1
  if score == len(least_integer_coefs):
    break
  else:
    integer +=1
leastIntegerMols = []
for i in least_integer_coefs:
  leastIntegerMols.append(int(round(integer*i)))
```

```python
leastIntegerMols
```

```python
elements = ['C', 'H', 'Cl'] # Re-initializing elements with the expected list from previous user input

leastIntegerMols = [1, 2, 1]


empirical_formula = ''

time = 0

while time < len(elements):
  empirical_formula += elements[time]
  empirical_formula += str(leastIntegerMols[time])
  time +=1

print(f'the empirical formula is {empirical_formula}')
```

```python
at_masses = []

for i in elements:
  at_masses.append(atomic_masses[i])

molar_mass_emp = 0

init = 0
while init < len(at_masses):
  molar_mass_emp += at_masses[init]*leastIntegerMols[init]
  init +=1

print(f'the empirical molar mass of the molecule is {molar_mass_emp}')
```

```python
molar_mass_cmpd = float(input('Molar mass of compound: '))

ratio = round(molar_mass_cmpd/molar_mass_emp)


#molecular formula

molecular_formula = ''

index = 0

while index < len(elements):
  molecular_formula += elements[index]
  molecular_formula += str(int(ratio*leastIntegerMols[index]))
  index +=1

print(f'The Molecular formula is{molecular_formula}')
```
